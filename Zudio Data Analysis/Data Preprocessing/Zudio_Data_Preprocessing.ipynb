```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Zudio_sales_data.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
# Convert Date Columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Store Open Date'] = pd.to_datetime(df['Store Open Date'])
```

```python
df.isnull().sum()
```

```python
# Handle Missing Values
# Numerical – fill with mean
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=['Contact Information'], inplace=True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
# Feature Engineering
df['Order_Year'] = df['Order Date'].dt.year
```

```python
df['Order_Month'] = df['Order Date'].dt.month
```

```python
df['Order_Day'] = df['Order Date'].dt.day
```

```python
df.head()
df.to_csv("Zudio_cleaned_data.csv", index=False)
```
